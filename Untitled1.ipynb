```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/used_cars.csv')
display(df.head())
```

### Cleaning `fuel_type` Column

```python
# Replace inconsistent entries with NaN and then fill nulls with the mode
df['fuel_type'] = df['fuel_type'].replace(['–', 'not supported'], np.nan)
mode_fuel_type = df['fuel_type'].mode()[0]
df['fuel_type'] = df['fuel_type'].fillna(mode_fuel_type)
print(f"Fuel type after cleaning and filling:")
print(df['fuel_type'].value_counts())
print(f"Nulls in fuel_type after cleaning: {df['fuel_type'].isnull().sum()}")
```

### Cleaning `accident` Column

```python
# Fill null values in 'accident' with the mode
mode_accident = df['accident'].mode()[0]
df['accident'] = df['accident'].fillna(mode_accident)
print(f"Accident status after filling nulls:")
print(df['accident'].value_counts())
print(f"Nulls in accident after cleaning: {df['accident'].isnull().sum()}")
```

### Handling `clean_title` Column

```python
# The 'clean_title' column has predominantly 'Yes' and NaN values.
# Since it doesn't offer much variability, we'll drop it.
print(f"Columns before dropping 'clean_title': {df.shape[1]}")
df.drop('clean_title', axis=1, inplace=True)
print(f"Columns after dropping 'clean_title': {df.shape[1]}")
```

### Converting `milage` and `price` to Numeric

```python
# Clean and convert 'milage' column
df['milage'] = df['milage'].str.replace(',', '').str.replace(' mi.', '', regex=False).astype(float)
print(f"'milage' column type: {df['milage'].dtype}")

# Clean and convert 'price' column
df['price'] = df['price'].str.replace('$', '', regex=False).str.replace(',', '').astype(float)
print(f"'price' column type: {df['price'].dtype}")

display(df[['milage', 'price']].head())
```

### Removing Duplicate Rows

```python
print(f"Number of rows before removing duplicates: {len(df)}")
df.drop_duplicates(inplace=True)
print(f"Number of rows after removing duplicates: {len(df)}")
```

### Handling Null Values

```python
print("Number of null values per column:")
print(df.isnull().sum())
```

### Getting Unique Values

```python
for col in df.columns:
    num_unique = df[col].nunique()
    if num_unique < 50:  # Display unique values for columns with less than 50 unique entries
        print(f"\nUnique values for column '{col}' ({num_unique} unique entries):\n{df[col].unique()}")
    else:
        print(f"\nColumn '{col}' has too many unique values to display ({num_unique} unique entries).")
```

### 1. Expanded Feature Selection for Price Prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

features_expanded = ['model', 'model_year', 'brand', 'milage', 'fuel_type', 'engine', 'transmission', 'ext_col', 'int_col', 'accident']
target_expanded = 'price'

X_expanded = df[features_expanded]
y_expanded = df[target_expanded]

print("Expanded features selected for price prediction:")
display(X_expanded.head())
```

### 2. One-Hot Encoding for Expanded Categorical Features

We need to convert all categorical columns in our expanded feature set into numerical representations for the linear regression model.

```python
categorical_features_expanded = ['model', 'brand', 'fuel_type', 'engine', 'transmission', 'ext_col', 'int_col', 'accident']

# Apply one-hot encoding
X_expanded_encoded = pd.get_dummies(X_expanded, columns=categorical_features_expanded, drop_first=True)

print("Shape of encoded expanded features:", X_expanded_encoded.shape)
print("Encoded expanded features head:")
display(X_expanded_encoded.head())
```

### 3. Data Splitting (Expanded Model)

Split the data into training and testing sets for the expanded model.

```python
# Split data into training and testing sets
X_train_expanded, X_test_expanded, y_train_expanded, y_test_expanded = train_test_split(X_expanded_encoded, y_expanded, test_size=0.2, random_state=42)

print(f"Training set size for expanded model: {len(X_train_expanded)} samples")
print(f"Testing set size for expanded model: {len(X_test_expanded)} samples")
```

### 4. Model Training (Expanded Model)

Train the Linear Regression model with the expanded set of features.

```python
# Initialize and train the Linear Regression model
model_expanded_price_pred = LinearRegression()
model_expanded_price_pred.fit(X_train_expanded, y_train_expanded)

print("Expanded Linear Regression model for price prediction trained successfully.")
```

### 5. Model Evaluation (Expanded Model)

Let's check the performance of the new model on the test data.

```python
y_pred_expanded_test = model_expanded_price_pred.predict(X_test_expanded)

mae_expanded = mean_absolute_error(y_test_expanded, y_pred_expanded_test)
r2_expanded = r2_score(y_test_expanded, y_pred_expanded_test)

print(f"Mean Absolute Error (MAE) for expanded model: {mae_expanded:.2f}")
print(f"R-squared (R2) Score for expanded model: {r2_expanded:.2f}")
```
